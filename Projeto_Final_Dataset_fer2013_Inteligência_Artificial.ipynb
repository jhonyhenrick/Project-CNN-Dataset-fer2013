```python
# Importação das bibliotecas necessárias
import tensorflow as tf
import gdown
import zipfile
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Download e descompactação do arquivo zip
zip_id = '1fo52iPHD-LggYu-9W5EPrTS8oLEY6-_h'
gdown.download(id=zip_id, output='archive.zip', quiet=False)
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/fer2013')

# Caminho base para as imagens descompactadas
dataset_path = '/content/fer2013/'

# Carregar imagens e labels
images = []
labels = []
for subdir, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            file_path = os.path.join(subdir, file)
            image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
            if image is not None:
                images.append(image)
                label = subdir.split('/')[-1]
                labels.append(label)

# Conversão para numpy arrays
images = np.array(images)
labels = np.array(labels)

# Normalização e redimensionamento
images = images / 255.0
images = images.reshape(-1, 48, 48, 1)

# Conversão de labels para formato categórico
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)
labels_categorical = to_categorical(labels_encoded)

# Divisão dos dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(images, labels_categorical, test_size=0.2, random_state=42)

# Construção do modelo
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(np.unique(labels)), activation='softmax')
])

# Configuração do otimizador com taxa de aprendizagem personalizada
opt = tf.keras.optimizers.Adam(learning_rate=0.001)

# Compilação do modelo com o otimizador personalizado
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
print(model.summary())

# Definição do callback de parada antecipada
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Treinamento do modelo
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test), batch_size=64, callbacks=[early_stopping])

# Plotagem dos gráficos de perda e acurácia
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Perda de Treinamento')
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda de Treinamento vs Validação')
plt.xlabel('Época')
plt.ylabel('Perda')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Acurácia de Treinamento')
plt.plot(history.history['val_accuracy'], label='Acurácia de Validação')
plt.title('Acurácia de Treinamento vs Validação')
plt.xlabel('Época')
plt.ylabel('Acurácia')
plt.legend()

plt.tight_layout()
plt.show()

# Salvar o modelo no formato recomendado
model.save('/content/modelo_fer2013.keras')
```

```python
# Avaliar o modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
# Carregar o modelo treinado
model = load_model('/content/modelo_fer2013.keras')

# Selecionar uma imagem aleatória do conjunto de teste
img_index = np.random.randint(0, X_test.shape[0])
sample_image = X_test[img_index]
sample_label = y_test[img_index]

# Fazer a previsão
pred_probabilities = model.predict(sample_image.reshape(1, 48, 48, 1))
pred_label = np.argmax(pred_probabilities, axis=1)

# Função para converter rótulos numéricos de volta para nomes de emoções
def get_emotion(label):
    label_map = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}
    return label_map[label]

# Exibir a imagem
plt.imshow(sample_image.reshape(48, 48), cmap='gray')
plt.title(f'Predicted Emotion: {get_emotion(pred_label[0])}\nTrue Emotion: {get_emotion(np.argmax(sample_label))}')
plt.axis('off')
plt.show()
```
